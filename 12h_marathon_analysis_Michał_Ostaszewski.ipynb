```python
df['Altitude Gain'] = df['Altitude Gain'].astype(str).str.replace('m', '').astype(float).astype(int) # Convert to float first to handle decimals and then to integer
df.head(5)
```

```python
# Lista kolumn do konwersji na float (zastąp rzeczywistymi nazwami kolumn)
columns_to_convert = ['Avg Power', 'Avg Cadence', 'Avg Torque', 'Avg HR', 'Max HR',
                      'Avg Gradient', 'Intensity', 'Avg Speed', 'Work in KJ',
                      'Distance', 'Work per km', 'Avg Power / Avg HR', 'Avg w/kg',
                      'Altitude Gain', 'Ambient Temperature', 'Body Temperature']

# Czyszczenie danych i zamiana na float
for column in columns_to_convert:
    # Usunięcie spacji, znaków specjalnych i zamiana przecinków na kropki
    # Check if the column has a string type before using the .str accessor
    if df[column].dtype == 'object':
        df[column] = df[column].str.replace(',', '.')

    # Konwersja na float, ignorując błędy
    df[column] = pd.to_numeric(df[column], errors='coerce')

# Wyświetlenie przekształconych danych
# df
```
